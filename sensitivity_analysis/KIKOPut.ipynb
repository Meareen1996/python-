```python
from KIKO import KIKOPut
```

### How Strike Price Affects Call Option Price

```python
S=100
L=90
U=120
R=80
T=1
n=50
sigma=0.2
r=0.05
M=1000

price=[]
param=[]

for K in range(10,200,10):
    option = KIKOPut(S, K, L, U, R, T, n, sigma, r, M)
    C = option.payoff()
    price.append(C)
    param.append(K)

plt.plot(param,price)
# naming the x axis
plt.xlabel('Strike Price')
# naming the y axis
plt.ylabel('Option Price')
plt.title(title+' Option')
```
